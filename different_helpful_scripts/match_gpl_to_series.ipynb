```python
import csv

def read_series():
    
    series = []
        
    with open('series_vs_platforms.csv','r') as tsvin:
        tsvin = csv.reader(tsvin, delimiter='\t')
        indexes = []
        
        for (i, row) in enumerate(tsvin):
            if i == 0:
                indexes = [row.index(j) for j in ['Series']]
            if row[indexes[0]] != 'Series':
                series.append(row[indexes[0]]) 
            
    return series    
                        

series = read_series()
```

```python
len(set(series))
```

```python
def read_series_from_server_file(file_name):
    lines = [line.rstrip('\n') for line in open(file_name)]
    return lines

series_server = read_series_from_server_file('series.txt')
```

```python
import re

def get_only_server_gse(lines):
    for i in range(0,len(lines),1):
        lines[i] = re.sub('(.*/)*', "", lines[i])
        lines[i] = re.sub('_series_matrix.txt.gz', "", lines[i])
        lines[i] = re.sub('\\.', "", lines[i])
        lines[i] = re.sub('-.*', "", lines[i])
    return lines

gse_server = get_only_server_gse(series_server)
```

```python
len(set(gse_server))
```

```python
# intersection_gse = set(gse_server).intersection(set(series))
intersection_gse = set(gse_server) & (set(series))
len(intersection_gse)
```

```python
gse_without_gpl = set(gse_server) ^ set(series)
print(len(gse_without_gpl))
gse_without_gpl
```

```python
with open('gse_without_gpl.txt','w') as f:
    f.write(str(gse_without_gpl))
```

```python
import csv


def read_platforms():
    
    gpl = []
        
    with open('rest_series_vs_gpl.csv','r') as tsvin:
        tsvin = csv.reader(tsvin, delimiter='\t')
        indexes = []
        
        for (i, row) in enumerate(tsvin):
            if i == 0:
                indexes = [row.index(j) for j in ['Platforms', 'Reason']]
            if row[indexes[0]] != 'Platforms' and row[indexes[0]] != '_' \
                                   and row[indexes[1]] != 'not rat, mouse, human ?':
                p = re.sub(' *', "", row[indexes[0]])    
                gpl.append(p) 
            
    return gpl    
                        

gpl = read_platforms()
```

```python
len(gpl)
```

```python
def read_platforms_from_server_file(file_name):
    lines = [line.rstrip('\n') for line in open(file_name)]
    return lines

platform_server = read_platforms_from_server_file('platforms.txt')
```

```python
import re

def get_only_server_gpl(lines):
    for i in range(0,len(lines),1):
        lines[i] = re.sub('(.*/)*', "", lines[i])
        lines[i] = re.sub('.annot.gz', "", lines[i])
        lines[i] = re.sub(' *', "", lines[i])
#         lines[i] = re.sub('\\.', "", lines[i])
#         lines[i] = re.sub('-.*', "", lines[i])
    return lines

gpl_server = get_only_server_gpl(platform_server)
len(gpl_server)
```

```python
len(set(gpl_server))
```

```python
intersection_gpl = set(gpl_server) & (set(gpl))
len(intersection_gpl)
```

```python
set(intersection_gpl)
```

```python
rest_gpl = set(gpl) - set(intersection_gpl)
print(len(rest_gpl))
rest_gpl
```

```python
from ftplib import FTP
from tqdm import trange
import csv



def write_dict_to_file(dict_for_download, file_name):
        
    with open(file_name, 'w') as csvout:
        csvout = csv.writer(csvout, delimiter='\t')
        rows = []
        rows.append(['key', 'path'])
        for key, value in dict_for_download.items():
            rows.append([key, value['path']])  
        csvout.writerows(rows)

        

def read_file(file_name):
    lines = [line.rstrip('\n') for line in open(file_name)]
    return lines        
        

    
def get_platforms_from_web(): 
    platforms_organism = read_file("rest_gpl.txt")

    ftp = FTP('ftp.ncbi.nlm.nih.gov')
    ftp.login()

    ftp.cwd('/geo/platforms/')
    platforms_list = ftp.nlst()

    annot = {}

    for i in trange(0,len(platforms_list),1):
        print('current platform: ', platforms_list[i], ' with i: ', i)
        
        ftp.cwd('/geo/platforms/' + platforms_list[i] + '/')
        gpl_list = ftp.nlst()

        for j in range(0,len(gpl_list),1):           
            path = platforms_list[i] + "/" + gpl_list[j] + "/annot/"
            if gpl_list[j] in platforms_organism:
                annot[gpl_list[j]] = {'path': path}
           
        print("__________________________________________________________________")
    
    return annot



annot = get_platforms_from_web()
print(len(annot))
write_dict_to_file(annot, "list_gpl.csv")
```

```python
def add_url_to_file(read_file, write_file):
    
    rows = []   
        
    with open(write_file, 'w') as csvout, open(read_file,'r') as tsvin:
        tsvin = csv.reader(tsvin, delimiter='\t')
        csvout = csv.writer(csvout, delimiter='\t')
        indexes = []
        
        for (i, row) in enumerate(tsvin):
            print(row)
            if i == 0:
                indexes = [row.index(j) for j in ['key', 'path']]
                rows.append(['key', 'path', 'url'])
        
            else:
                url = 'https://ftp.ncbi.nlm.nih.gov/geo/platforms/' + row[indexes[1]] + \
                        row[indexes[0]] + '.annot.gz'
                rows.append([row[indexes[0]], row[indexes[1]], url])
        
        csvout.writerows(rows)
        
        
add_url_to_file("list_gpl.csv", "list_url_gpl.csv")
```

```python
import subprocess
import csv


def download_data(file_name): 
    
    with open(file_name,'r') as tsvin:
        tsvin = csv.reader(tsvin, delimiter='\t')
        indexes = []
        
        for (i, row) in enumerate(tsvin):
            if i == 0:
                indexes = [row.index(j) for j in ['key', 'path', 'url']]
                
            else:    
                try:
                    directory = "../geo/platforms/" + row[indexes[1]]
                    subprocess.call(["mkdir", "-p", directory])
                    subprocess.call(["wget", "-c", "-P", directory, row[indexes[2]]])
                except:
                    pass

            print(row[indexes[0]], ', url: ', row[indexes[2]], ' with i: ', i)

        
        
download_data("list_url_gpl.csv") # only i = 14,24,26,30,32,34,36,37 were downloaded
```
